```python
import pandas as pd
```

```python
#Reading CSV file 
Torontopath='http://cocl.us/Geospatial_data'
dfPCLL=pd.read_csv(Torontopath)
dfPCLL=dfPCLL.rename(columns={'Postal Code':'PostalCode'})
dfPCLL.head()
```

```python
#Reading toronto wiki burough CSV file 
path='./TorontoWiki.csv'
df=pd.read_csv(path)
df=df.drop(columns='Unnamed: 0')
df.head()
```

```python
dfPCLL.shape
type(df.iloc[0][0])
```

```python
#se agregan las nuevas columnas de latitud y longitud
df['Latitude']=None
df['Longitude']=None
df
```

```python
# agregar los datos de lat y ling segun el codigo postal del barrio
for i in df.index:
        for j in dfPCLL.index:
            #print(df['Postal Code'][i])
            #print(dfPCLL['Postal Code'][j])
            if (df['PostalCode'][i]==dfPCLL['PostalCode'][j]):
                df['Latitude'][i]=dfPCLL['Latitude'][j]
                df['Longitude'][i]=dfPCLL['Longitude'][j]
df 
```

```python
#botar el df en un archivo que pueda ser leido desde un archivo
df.to_csv('TorontoBorLatLong.csv')     
```
